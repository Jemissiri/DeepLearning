<h1>Loading and visualizing a variant of MNIST digits </h1>

Optical character recognition, and the simpler digit recognition task, has been around for many years as one of the most popular applications of Machine Learning. In the following we are going to compare empirically a set of digit recognizers using the companion data set with 5000 examples from the data set MNIST available from http://yann.lecun.com/exdb/mnist/.


To start with, let us load digits.mat and observe the loaded variables. Notice that the X matrix has the 5000 digits, one in each row. Each row represents an image with a 20x20 digit, hence X has 5000x400 elements.

```python
import numpy as np
import scipy.io
import matplotlib.pyplot as plt

raw=scipy.io.loadmat('digits.mat')
X = raw['X']
print (X.shape)
y = raw['y']
```

```python
def displaySamples(X = None, sample_width = None):
    '''
    Display a series of sample bitmaps in X, with sample_width
    '''

    # Set sample_width to a square if not passed as arg
    m,n = X.shape
    if (sample_width is None) or len(sample_width)<=0:
        sample_width = int(np.round(np.sqrt(n)))
    sample_height = int(n / sample_width)

    for i in range(m):
        a = np.reshape(X[i,:], (sample_height, sample_width), order="F" )
        # print(i, a.shape)
        h = plt.imshow(a)
        plt.axis('off')
        # Gray Image
        plt.gray()
        plt.show()
```

Now let's visualize a given no_of_digits, say 4

```python
m = X.shape[0]  # 5000
N = 4

# Randomly select N samples to display
rand_indices = np.random.permutation(m)
sri = rand_indices[np.arange(N)]
sel = X[sri,:]
displaySamples(sel)
print (y[sri])
```
